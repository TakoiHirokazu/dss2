```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
from tqdm import tqdm_notebook as tqdm
import sys
from sklearn.preprocessing import LabelEncoder
sys.path.append("../src/")
from logger import setup_logger, LOGGER
from trainer import train_lgbm
from util_tool import reduce_mem_usage
%matplotlib inline
pd.set_option('display.max_columns', 300)
```

```python
# ==================
# Constant
# ==================
TRAIN_PATH = "../input/train.csv"
TEST_PATH = "../input/test.csv"
USER_PATH = "../input/user_x_anime.csv"
```

```python
# =====================
# Settings
# =====================
SAVE_PATH = "../output/fe/fe019.feather"
```

```python
train = pd.read_csv(TRAIN_PATH)
test = pd.read_csv(TEST_PATH)
```

```python
train.head()
```

```python
cat_cols = ['Source',"Rating"]
for c in cat_cols:
    train.loc[~train[c].isin(test[c].unique()),c] = np.nan
    test.loc[~test[c].isin(train[c].unique()),c] = np.nan
train = pd.concat([train[cat_cols],test[cat_cols]]).reset_index(drop=True)
```

```python
for c in cat_cols:
    train[c] = train[c].astype(str)
    le = LabelEncoder()
    train[c] = le.fit_transform(train[c])
```

```python
train = reduce_mem_usage(train)
train.to_feather(SAVE_PATH)
```
