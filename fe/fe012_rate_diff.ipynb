```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
from tqdm import tqdm_notebook as tqdm
import itertools
import sys
sys.path.append("../src/")
from logger import setup_logger, LOGGER
from trainer import train_lgbm
from util_tool import reduce_mem_usage
%matplotlib inline
pd.set_option('display.max_columns', 300)
```

```python
# ==================
# Constant
# ==================
TRAIN_PATH = "../input/train.csv"
TEST_PATH = "../input/test.csv"
USER_PATH = "../input/user_x_anime.csv"
```

```python
# =====================
# Settings
# =====================
SAVE_PATH = "../output/fe/fe012.feather"
```

```python
train = pd.read_csv(TRAIN_PATH)
test = pd.read_csv(TEST_PATH)
```

```python
fe007 = pd.read_feather("../output/fe/fe007.feather")
```

```python
numerical_cols = ["Members", "Favorites","Watching","Completed","On-Hold","Dropped","Plan to Watch"]
train = pd.concat([train[numerical_cols],test[numerical_cols]]).reset_index(drop=True)
train = pd.concat([train,fe007],axis=1)
```

```python
cols = train.columns
new_cols = []
for c1, c2 in itertools.combinations(cols, 2):
    train[f"{c1}_{c2}_diff"] = train[c1] - train[c2]
    new_cols.append(f"{c1}_{c2}_diff")
    train[f"{c1}_{c2}_rate"] = train[c1] / train[c2]
    new_cols.append(f"{c1}_{c2}_rate")
```

```python
train = reduce_mem_usage(train)
train[new_cols].to_feather(SAVE_PATH)
```
